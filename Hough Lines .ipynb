```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('handmobile.jpg')
image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

### Perform Edge Detection

```python
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

# define params for canny edge detection
lower = 50
higher = 100

edges = cv2.Canny(gray, lower, higher)

plt.imshow(edges, cmap='gray')
```

### Find line using Hough transform

```python
# Define Hough params
rho = 1
theta = np.pi/180
threshold = 60
min_line_length = 80
max_line_gap = 5
```

```python
lines = cv2.HoughLinesP(edges, rho, theta, threshold, np.array([]), min_line_length, max_line_gap)
```

```python
line_image = np.copy(image_copy)
```

```python
# Iterate over the output lines and draw lines on the image copy
for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(line_image, (x1, y1), (x2, y2), (255, 0, 0), 5)

plt.imshow(line_image)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Read in an image and convert to RGB
image = cv2.imread('road3.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
%matplotlib inline
```

```python
# Convert image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray, cmap='gray')
```

```python
# Peform edge detection with Canny
low_threshold = 90
high_threshold = 180
edges = cv2.Canny(gray, low_threshold, high_threshold)
# ---------------------------------------------------------- #
```

```python
plt.imshow(edges, cmap='gray')
```

```python
## TODO: Define the Hough transform parameters 
## so that lane lines are the only ones detected
rho = 1
theta = np.pi/180
threshold = 50
min_line_length = 120
max_line_gap = 8
```

```python
line_image = np.copy(image) #creating an image copy to draw lines on

# Run Hough on the edge-detected image
lines = cv2.HoughLinesP(edges, rho, theta, threshold, np.array([]),
                        min_line_length, max_line_gap)
```

```python
# Iterate over the output "lines" and draw lines on the image copy
for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),3)

# ---------------------------------------------------------- #
```

```python
plt.imshow(line_image)
```
