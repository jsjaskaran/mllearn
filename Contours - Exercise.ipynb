```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
%matplotlib inline
```

```python
# Read image
image = cv2.imread('hand-06.jpg')

# make copy
image_copy = np.copy(image)

# change to RGB
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

### Produce a binary image for finding contours

```python
# convert to grayscale
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

# create a binary threshold image
retval, binary = cv2.threshold(gray, 240, 255, cv2.THRESH_BINARY_INV)

plt.imshow(binary, cmap='gray')
```

### Find and draw the contours

```python
# Find the contours from threshold image
# args - binary image, contour retrieval mode - Tree here, contour approx method
# hierarchy - useful if you have many contours nested within one another
retval, contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Draw all contours on a copy of the original image
image_copy2 = np.copy(image_copy)
# -1 to display all contours, color, size
all_contours = cv2.drawContours(image_copy2, contours, -1, (0, 255, 0), 5)

plt.imshow(all_contours)
```

```python
# Assignment
# Read image
image = cv2.imread('thumbs-up-down.jpg')

# make copy
image_copy = np.copy(image)

# change to RGB
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
# convert to grayscale
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

# create a binary threshold image
retval, binary = cv2.threshold(gray, 240, 255, cv2.THRESH_BINARY_INV)

plt.imshow(binary, cmap='gray')
```

```python
retval, contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Draw all contours on a copy of the original image
image_copy2 = np.copy(image_copy)
# -1 to display all contours, color, size
all_contours = cv2.drawContours(image_copy2, contours, -1, (0, 255, 0), 2)

plt.imshow(all_contours)
```

```python
# Select the contour at index= 0 from a list
selected_contour = contours[0]
```

```python
# Draw the first contour (index = 0)
image_copy3 = np.copy(image_copy)
contour_image = cv2.drawContours(image_copy3, [selected_contour], 0,  (0,255,0), 3)

plt.imshow(contour_image)
```
